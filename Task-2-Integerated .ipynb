```python
orders["ORDER_TOTAL"]=np.where(orders.ORDER_TOTAL.isnull(),orders.ORDER_TOTAL.mean(),orders.ORDER_TOTAL)
orders["DISCOUNT"]=np.where(orders.DISCOUNT.isnull(),orders.DISCOUNT.mean(),orders.DISCOUNT)
orders["DELIVERY_STATUS"]=np.where(orders.DELIVERY_STATUS.isnull(),orders.DELIVERY_STATUS.mode(),orders.DELIVERY_STATUS)
```

```python
orders["Total_Revenue"]=(orders.ORDER_TOTAL-orders.DISCOUNT)
```

```python
### Merging both the tables
merged=pd.merge(cust,orders,on="CUSTOMER_KEY",how='inner')
merged
```

### Q1. Number of orders by month based on order status (Delivered vs. canceled vs. etc.) - Split of order status by month

```python

pivot_table = pd.pivot_table(merged,
                              index=merged.ORDER_DATE.dt.month_name(),
                              columns='ORDER_STATUS',
                              values='ORDER_NUMBER',
                              aggfunc='count')
pivot_table
```

### Q3. Month-on-month growth in OrderCount and Revenue (from Novâ€™15 to Julyâ€™16)

```python
# Filter data 
filtered_data = merged[(merged['ORDER_DATE'] >= '2015-11-01') & (merged['ORDER_DATE'] <= '2016-07-31')]

# Group data by month and calculate total order count and revenue for each month
monthly_summary = filtered_data.groupby(filtered_data['ORDER_DATE'].dt.month_name()).agg(OrderCount=('ORDER_NUMBER', 'count'), Revenue=('Total_Revenue', 'sum'))

# Calculate month-on-month growth in order count and revenue
monthly_summary['OrderCount_Growth'] = monthly_summary['OrderCount'].pct_change() * 100
monthly_summary['Revenue_Growth'] = monthly_summary['Revenue'].pct_change() * 100


print(monthly_summary.sort_values(by=['OrderCount_Growth','Revenue_Growth']))
```

### Q6. Write user defined function which take inputs as location & month, and the output is total_order value and number of orders by Gender, Delivered Status for given location & month. Test the code with different options


```python
def total_order(Location,Month):
    # Filter the DataFrame by location and month
    filtered_dataframe=merged[(merged['Location']==Location)&(merged['ORDER_DATE'].dt.month_name())]
    # Create a pivot table
    pivot_table=pd.pivot_table(filtered_dataframe,index='Gender',columns='DELIVERY_STATUS',values=['ORDER_TOTAL','ORDER_NUMBER'],aggfunc={'ORDER_TOTAL': 'sum', 'ORDER_NUMBER': 'count'})
    return pivot_table
```
