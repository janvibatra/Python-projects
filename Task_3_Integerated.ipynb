```python
orders["ORDER_TOTAL"]=np.where(orders.ORDER_TOTAL.isnull(),orders.ORDER_TOTAL.mean(),orders.ORDER_TOTAL)
orders["DISCOUNT"]=np.where(orders.DISCOUNT.isnull(),orders.DISCOUNT.mean(),orders.DISCOUNT)
orders["DELIVERY_STATUS"]=np.where(orders.DELIVERY_STATUS.isnull(),orders.DELIVERY_STATUS.mode(),orders.DELIVERY_STATUS)
```

```python
orders["Total_Revenue"]=(orders.ORDER_TOTAL-orders.DISCOUNT)
```

```python
### Merging both the tables
merged=pd.merge(cust,orders,on="CUSTOMER_KEY",how='inner')
merged
```

### Q1. Total Revenue, total orders by each location

```python
merged.groupby("Location").agg(TOTAL_Order=('ORDER_NUMBER', 'count'), Revenue=('Total_Revenue', 'sum'))
```

### Q2. Total revenue, total orders by customer gender


```python
merged.groupby("Gender").agg(TOTAL_Order=('ORDER_NUMBER', 'count'), Revenue=('Total_Revenue', 'sum'))
```

### Q3. Which location of customers cancelling orders maximum?

```python
order_cancelled=merged[merged.ORDER_STATUS=='Cancelled'] #filetring data
order_cancelled.groupby("Location")["CUSTOMER_KEY"].count().sort_values(ascending=False).head(1) #summarising on location
```

### Q4. Total customers, Revenue, Orders by each Acquisition channel

```python
merged.groupby("Acquired Channel").agg(Total_customers=('CUSTOMER_KEY','count'),TOTAL_Order=('ORDER_NUMBER', 'count'), Revenue=('Total_Revenue', 'sum'))
```

### Q6. Write User Defined Function which can take input data set which create two data sets with numerical variables and categorical variables separately

```python
def variable(data):
    cont=data.select_dtypes(include=['float','int'])
    cat=data.select_dtypes(include='object')
    return cont,cat

cont, cat = variable(cust)
print("Numerical variables:")
print(cont)
print("\nCategorical variables:")
print(cat)
```

### Q7. Write User Defined Function which can take input as any variable and output as follows. Test them on different columns
### a. For categorical variables, the output will be number of distinct categories with frequency, number of missings, percentage of missings
### b. For numerical variables, the output will be no.of missings, percentage of missings, minimum, maximum, sum, mean, median, variance, Coeficient of variation, 25th Percentile and 75th Percentile values

```python
def columns(data,var):
    if data[var].dtype == 'O':
        cat_fre=data[var].value_counts()
        num_cat=len(cat_fre)
    
        missing_count = data[var].isnull().sum()
        total_count = len(data)
        missing_percentage = (missing_count / total_count) * 100
    
        # Display results
        print("Analysis for variable:", var)
        print("Number of distinct categories:", num_cat)
        print("Distinct categories with frequency:")
        print(cat_fre)
        print("Number of missing values:", missing_count)
        print("Percentage of missing values:",missing_percentage)
        
    elif data[var].dtype=='float64':
        # For numerical variables
        missing_count = data[var].isnull().sum()
        total_count = len(data)
        missing_percentage = (missing_count / total_count) * 100
        minimum = data[var].min()
        maximum = data[var].max()
        total_sum = data[var].sum()
        mean = data[var].mean()
        median = data[var].median()
        variance = data[var].var()
        coeff_variation = (np.std(data[var]) / mean) * 100
        percentile_25 = data[var].quantile(0.25)
        percentile_75 = data[var].quantile(0.75)
        
        print("Analysis for numerical variable:", var)
        print("Number of missing values:", missing_count)
        print("Percentage of missing values: {:.2f}%".format(missing_percentage))
        print("Minimum:", minimum)
        print("Maximum:", maximum)
        print("Sum:", total_sum)
        print("Mean:", mean)
        print("Median:", median)
        print("Variance:", variance)
        print("Coefficient of variation: {:.2f}%".format(coeff_variation))
        print("25th Percentile:", percentile_25)
        print("75th Percentile:", percentile_75)
```
